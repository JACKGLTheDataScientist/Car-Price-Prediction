#### Notes - Cleaning Task 1 

Wheel - Value Change - Left Wheel to 'Left-hand drive' \
Doors - Value & Type Change - '04-May' = 4, '02-May' =2, '>5'= '5' 

```python
# Value change in Wheel - 'Left wheel' to 'Left-hand drive'
lst = ['Left wheel']
data['Wheel'] = np.where(data['Wheel'].isin(lst), 'Left-hand drive', data.Wheel)
data['Wheel'].head(10)
```

```python
# Chaning Doors Value 
# Changing the following: '04-May' = 4, '02-May' =2, '>5'= '5' 

# '04-May' = 4,
lst_1 = ['04-May']
data['Doors'] = np.where(data['Doors'].isin(lst_1), '4-Door', data.Doors)
data.head(5)

# '02-May' =2
lst_2 = ['02-Mar']
data['Doors'] = np.where(data['Doors'].isin(lst_2), '2-Door', data.Doors)

# >5'= '5
lst_3 = ['>5']
data['Doors'] = np.where(data['Doors'].isin(lst_3), '5-Doors', data.Doors)

# Checking the values correctly changed in the Doors variable
data['Doors'].value_counts()

# Change to type int64
data['Doors']
```

##### Data type and conversions 
Object -Data type used for strings \
int64 - Used for intergers (whole No.) \
float64 - Used for floating point numbers (decimals/fractions) \
bool - boolean (True and False) \
datetime64 - date and time values \
timedelta - represents difference between two datatimes \
category - values that take one out of limited number of available options
